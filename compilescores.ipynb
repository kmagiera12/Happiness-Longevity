```python
def compile_emotions(images=os.listdir('/Users/kamilamagiera/Documents/DA401/Celeb Photos')):
    images.sort()
    scores = []
    for i in images:
        try:
            img = cv2.imread(f'/Users/kamilamagiera/Documents/DA401/Celeb Photos/{i}')

            # storing the result
            result = DeepFace.analyze(img,
                                      actions = ['emotion'])
            scores.append(result["emotion"]['happy'])
        except:
            scores.append(101)
    return scores
```
